```python
import pickle
with open('rating.pickle', 'rb') as handle:
        goods = pickle.load(handle)
```

```python
dict(sorted(goods.items(), key=lambda item: item[1])[::-1])
```

```python

marks = []
nums = []
for k, v in goods.items():
  marks += [v[0]]
  nums += [v[1]]

import statistics

R = statistics.median(marks)
W = statistics.mean(nums)/20
print('R:', R, 'W:', W)
```

```python
len(marks)
```

```python
all_goods = []
for k, v in goods.items():
  m, n= v[0], v[1]
  rat = (R*W + m*n)/(W + n)
  all_goods += [(rat, k)]
```
