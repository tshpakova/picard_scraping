```python
from bs4 import BeautifulSoup, Comment
import requests

#id = "pagelet_forsale_island"

picard = 'https://www.picard.fr/'
r = requests.get(picard)
soup = BeautifulSoup(r.content, "html.parser")
```

```python
goods = {}

tags = soup.find_all('a', class_ = 'pi-Nav-panelTitle-link')
for t in tags:
  # section
  res = str(t).split('">')[1].split('</a>')[0]
  link = picard[:-1] + t.attrs["href"]
  #print(res, link)
  r = requests.get(link)
  soup_t = BeautifulSoup(r.content, "html.parser")

  tags2 = soup_t.find_all('a', class_ = 'pi-ProductCard-link js-gtm-ProductCard-link')
  for t2 in tags2:
    # produit

    link2 = picard[:-1] + t2.attrs['href']
    r2 = requests.get(link2)
    soup_t2 = BeautifulSoup(r2.content, "html.parser")
    tags3 = soup_t2.find_all('span', class_ = 'sr-only')
    for t3 in tags3:
      if 'sur' in str(t3):
        break

    rating = float(str(t3).split('only">')[1].split(' sur')[0])

    if rating:

      reviews = soup_t2.find_all('a', class_ = 'pi-Reviews-linksItem js-GoToReviews')
      num = int(str(reviews).split('Lire ')[1].split(' ')[0])

      if 'produits/' in link2:
        goods[link2.split('produits/')[1].split('.html')[0]] = [rating, num]
        print(res, link2.split('produits/')[1].split('.html')[0], rating, num)

      else:
         goods[link2.split('fr/')[1].split('.html')[0]] = [rating, num]
         print(res, link2.split('fr/')[1].split('.html')[0], rating, num)
```

```python
import pickle

with open('rating.pickle', 'wb') as handle:
                pickle.dump(goods, handle, protocol=pickle.HIGHEST_PROTOCOL)
```
